```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import json
import re
```

```python
subject_pattern = re.compile("(K|ALS)\d{2,3}(_t\d+)?")
```

```python
subject_pattern.search("testing on K309...").group()
```

```python
res = list()
subject = None
metrics = None
with open("/home/txdat/Documents/ALS02_LRF0_K_final2.txt", mode="r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line.startswith("testing on"):
            subject = subject_pattern.search(line).group()
        elif line.startswith("{'acc'"):
            if subject is not None:
                metrics = json.loads(line.replace("'", "\""))
                metrics["subject"] = subject
                res.append(metrics)
                subject = None
                
res = pd.DataFrame(res)
res
```
