```python
import numpy as np
import pandas as pd
import mne
import seaborn as sb
import matplotlib.pyplot as plt

from data_util import load_subjects, load_sample_data, load_subject_data, load_data
```

```python
data_dir = "../data/vin/Official"
```

```python
subjects_scenarios = load_subjects(data_dir)
print(f"found {len(subjects_scenarios)} subjects")

scenario_count = dict()
for _, scenarios in subjects_scenarios:
    for scenario in scenarios:
        if scenario in scenario_count:
            scenario_count[scenario] += 1
        else:
            scenario_count[scenario] = 1

plt.bar(scenario_count.keys(), scenario_count.values())
plt.xticks(rotation="vertical")
```

```python
%%time
scenarios = [
    "nâng tay trái",
    "nâng tay phải",
    "nâng chân trái",
    "nâng chân phải",
    "gật đầu",
    "lắc đầu",
    # "há miệng",
]

events = [
    "Thinking",
    # "Thinking and Acting",
    # "Resting",
    # "Typing",
]

subjects = sorted(
    [subject for subject, _ in subjects_scenarios if subject.startswith("K")]
)

# channels = [
#     "Fp1", "F7", "F3", "FC5", "FC1", "T7", "C3", "CP5", "CP1", "P7", "P3", "O1",  # left
#     "Fz", "Cz", "Pz", "Oz",  # center
#     "Fp2", "F8", "F4", "FC6", "FC2", "T8", "C4", "CP6", "CP2", "P8", "P4", "O2",  # right
# ]
channels = [
    "Fp1",
    "F7",
    "F3",
    "FC5",
    "FC1",
    "T7",
    "C3",
    "CP5",
    "CP1",
    "P7",
    "P3",
    "O1",  # left
    "Fz",
    "Cz",
    "Pz",
    "Oz",  # center
    "Fp2",
    "F8",
    "F4",
    "FC6",
    "FC2",
    "T8",
    "C4",
    "CP6",
    "CP2",
    "P8",
    "P4",
    "O2",  # right
]

print(f"using {len(channels)} channels")

window_size = 640  # @param
window_stride = 64  # @param

fmin = 1.0  # @param
fmax = 45.0  # @param

alignment = False  # @param {"type": "boolean"}

ds = load_data(
    data_dir,
    train_subjects=[subj for subj, _ in subjects_scenarios if subj.startswith("K")],
    val_subjects=None,
    test_subjects=None,
    window_size=window_size,
    window_stride=window_stride,
    scenarios=scenarios,
    events=events,
    channels=channels,
    fmin=fmin,
    fmax=fmax,
    resample=None,
    alignment=alignment,
)
```

```python
ds_info = ds.description
ds_info
```

```python
from collections import defaultdict
```

```python
trial_times = defaultdict(list)

for i, r in ds_info.iterrows():
    annot = pd.DataFrame(ds.datasets[i].raw.annotations)
    for ia, ra in annot.iterrows():
        if ra["description"] == "Thinking":
            trial_times[r["scenario"]].append(ra["duration"])
```

```python
for scenario, duration in trial_times.items():
    sb.distplot(duration, hist=False)
```
