```python
def covariance(x):
#     xm = x - x.mean(axis=2, keepdims=True)
#     c = (xm @ xm.transpose((0, 2, 1))) / (xm.shape[2] - 1)
    
#     return c
    return np.stack([np.cov(xi, rowvar=True) for xi in x])
```
