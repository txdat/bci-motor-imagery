```python
ds_info = ds.description
ds_info
```

```python
x = list()
y = list()
for i in ds_info[ds_info["subject"] == "K300"].index:
    x.append(ds.datasets[i].windows.get_data())
    y.extend(ds.datasets[i].y)
    
x = np.concatenate(x, axis=0)
y = np.array(y, dtype=int)

x.shape, y.shape
```

```python
c = np.stack([np.cov(xi, rowvar=True) for xi in x])
c.shape
```

```python
from sklearn.manifold import TSNE
from pyriemann.tangentspace import TangentSpace
```

```python
ts = TangentSpace(metric="riemann")

feat = ts.fit_transform(c)
```

```python
%%time
feat2d = TSNE(n_components=2).fit_transform(feat)
```

```python
epochs.plot_psd(average=False)
```

```python
freq_range = np.arange(8, 30)

power = mne.time_frequency.tfr_multitaper(epochs, freqs=freq_range, n_cycles=freq_range / 2.0, use_fft=True,
                                          return_itc=False, average=True, decim=2)
```

```python
power.plot(title="auto", mode="logratio")
```

```python
fig, axis = plt.subplots(1, 2, figsize=(12, 3))

power.plot_topomap(ch_type='eeg', fmin=8, fmax=12,
                   mode='logratio', axes=axis[0],
                   title='Alpha', show_names=True, show=False)
power.plot_topomap(ch_type='eeg', fmin=12, fmax=30,
                   mode='logratio', axes=axis[1],
                   title='Beta', show_names=True, show=False)
```

```python
power.plot_joint(timefreqs=((0.25, 10), (0.5, 10), (1.0, 10), (1.5, 10)))
```
