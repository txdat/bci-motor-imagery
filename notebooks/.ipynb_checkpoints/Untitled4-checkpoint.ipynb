```python
import numpy as np
import pandas as pd
import mne
import matplotlib.pyplot as plt
import seaborn as sb
import eegraph
```

```python
raw = mne.io.read_raw_edf("../data/vin/Official/K300/sample1/EEG.edf")
raw.set_montage("standard_1005")

raw.plot(n_channels=32, scalings="auto")
```

```python
pd.DataFrame(raw.annotations)
```

```python
G = eegraph.Graph()
G.load_data("../data/vin/Official/K300/sample1/EEG.edf")
```

```python
graphs, connectivity_matrix = G.modelate(window_size=2, connectivity='pearson_correlation')
```

```python
sb.heatmap(connectivity_matrix[42])
```

```python
G.visualize(graphs[42], 'graph_42')
```

```python
anno = pd.DataFrame(raw.annotations)
```

```python
du = int(2 * 128.0)

for i, r in anno[anno["description"] == "Thinking"].iterrows():
    onset = int(r["onset"] * 128.0)
    offset = int((r["onset"] + r["duration"]) * 128.0)
    
    for j in range(onset // du, offset // du + 1):
        G.visualize(graphs[j], f'graph_{i}_{j}')
```

```python
df = pd.read_csv("~/shit2.csv", header=None)
df
```

```python
dd = list()
for _, r in df.iterrows():
    x = r[0].split()
    dd.append((x[0], *(float(xi) for xi in x[1:])))
    
dd = pd.DataFrame(dd, columns=["subject", "acc2", "f12", "acc3", "f13", "acc1", "f11"])
dd
```

```python
dd["acc"] = 0
dd["f1"] = 0

for i, r in dd.iterrows():
    dd.loc[i, "acc"] = (r["acc1"] + r["acc2"] + r["acc3"]) / 3
    dd.loc[i, "f1"] = (r["f11"] + r["f12"] + r["f13"]) / 3
```

```python
dd
```

```python
# dd.loc[38, "acc"] = (dd.iloc[38]["acc2"] + dd.iloc[38]["acc3"]) / 2
# dd.loc[38, "f1"] = (dd.iloc[38]["f12"] + dd.iloc[38]["f13"]) / 2
dd.loc[39, "acc"] = (dd.iloc[39]["acc2"] + dd.iloc[39]["acc3"]) / 2
dd.loc[39, "f1"] = (dd.iloc[39]["f12"] + dd.iloc[39]["f13"]) / 2
```

```python
dd["acc"].mean(), dd["f1"].mean()
```
