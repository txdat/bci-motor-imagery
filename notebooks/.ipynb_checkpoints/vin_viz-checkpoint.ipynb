```python
import numpy as np
import pandas as pd
import mne
import seaborn as sb
import matplotlib.pyplot as plt

from data_util import load_sample_data, load_subject_data, load_data
```

```python
data_dir = "../data/vin/Official"
```

```python
ds = load_data(
    data_dir,
    scenarios=[
        "nâng tay trái",
        "nâng tay phải",
        "nâng chân trái",
        "nâng chân phải",
        "gật đầu",
        "lắc đầu",
    ],
    events=[
        "Thinking",
    ],
    return_raw=True,
)
```

```python
ds_info = ds.description
ds_info
```

```python
ds_info.groupby("scenario").count()
```

```python
from collections import defaultdict
```

```python
trial_times = defaultdict(list)

for i, r in ds_info.iterrows():
    annot = pd.DataFrame(ds.datasets[i].raw.annotations)
    for ia, ra in annot.iterrows():
        if ra["description"] == "Thinking":
            trial_times[r["scenario"]].append(ra["duration"])
```

```python
for scenario, duration in trial_times.items():
    sb.distplot(duration, hist=False)
```

```python
picks = mne.pick_channels(
    ch_names=ds.datasets[0].raw.info["ch_names"],
    include=[
        "Fp1", "F7", "F3", "FC5", "FC1", "T7", "C3", "CP5", "CP1", "P7", "P3", "O1",  # left
        "Fz", "Cz", "Pz", "Oz",  # center
        "Fp2", "F8", "F4", "FC6", "FC2", "T8", "C4", "CP6", "CP2", "P8", "P4", "O2",  # right
    ],
    ordered=True,
)
```

```python
epochs = list()

for i, r in ds_info[(ds_info["scenario"] == "nâng chân trái") & (ds_info["subject"] == "K320")].iterrows():
    raw = ds.datasets[i].raw
    annot = pd.DataFrame(raw.annotations)
    events, event_id = mne.events_from_annotations(raw)
    id_event = {_id: _name for _name, _id in event_id.items()}
    
    for j, event in enumerate(events):
        if id_event[event[-1]] != "Thinking":
            continue
        
        epochs.append(
            mne.Epochs(
                raw,
                event[None],
                event_id=event_id,
                tmin=0,
#                 tmax=annot.iloc[j]["duration"],
                tmax=6,
                proj=False,
                baseline=None,
                preload=True,
                verbose=False,
                picks=picks,
                event_repeated="drop",
                on_missing="ignore",
            )
        )
        
epochs = mne.concatenate_epochs(epochs)
```

```python
epochs.plot_psd()
```

```python
evoked = epochs.average()
```

```python
evoked.plot_topomap()
```
